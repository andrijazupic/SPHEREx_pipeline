# Data

```python
data = [
    {"source_id": 4012093916593061120, 
     "ra": 186.628469, 
     "dec": 30.647967},
    {"source_id": 1662622870104953472, 
     "ra": 205.250219, 
     "dec": 60.436094},
    {"source_id": 318381458887086976, 
     "ra": 25.998160, 
     "dec": 35.533183}
]

df = pd.DataFrame(data).astype({"source_id": "Int64"})
```

# Get SPHEREx spectrum

- Function get_spherex_data_calibrated queries SPHEREx data. 
- If calibrate=True, applies empirical aperture calibration.
- Saves a flattened CSV directly to the outdir and deletes ALL temporary pipeline folders.

```python
for index, row in df.iterrows():
    source_df = spxquery_get_spectrum_calibrated(source_id=row['source_id'],
                                                 ra=row['ra'], 
                                                 dec=row['dec'], 
                                                 calibrate=False, 
                                                 outdir="spxquery_results")
```

```python
for index, row in df.iterrows():
    source_df = spxquery_get_spectrum_calibrated(source_id=row['source_id'],
                                                 ra=row['ra'], 
                                                 dec=row['dec'], 
                                                 calibrate=True, 
                                                 outdir="spxquery_results_calib")
```
